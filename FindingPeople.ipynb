# Vemos el tamaño de batch óptimo

```python
import torch

def get_optimal_batch_size(image_size=640):
    # Obtener memoria total de la primera GPU
    total_memory = torch.cuda.get_device_properties(0).total_memory // (1024 ** 2)  # en MiB
    print(f"Total GPU memory: {total_memory} MiB")
    
    # Calcular batch size aproximado en función del tamaño de imagen y la memoria disponible
    if total_memory >= 16000:  # 16 GB o más
        return 32
    elif total_memory >= 8000:  # Entre 8 GB y 16 GB
        return 16
    elif total_memory >= 4000:  # Entre 4 GB y 8 GB
        return 8
    else:
        return 2  # Menos de 4 GB

optimal_batch_size = get_optimal_batch_size()
print(f"Optimal batch size: {optimal_batch_size}")
```
